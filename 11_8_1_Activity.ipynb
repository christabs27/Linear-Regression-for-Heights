#Step 5: Plot the temperature data over time.
* Run the following code block to visualize the daily low temperature in Melbourne, AU from 1981 to 1990.

````python
#Step 5

temperatures.plot()
plt.rcParams["figure.figsize"] = [7.50, 9.50]
plt.ylabel('Daily Minimum Temperature in C')
plt.title('Daily Mimimum Temperature in Melbourne, Australia: 1981 - 1990')
plt.xticks(rotation=90)
plt.show()
````

#Step 6: Plot the autocorrelation of the temperature each day with the temperature the day before.
